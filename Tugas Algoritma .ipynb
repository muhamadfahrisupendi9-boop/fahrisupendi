```
while True:
    print("\n=== MENU ===")
    print("1. A pangkat B")
    print("2. Hitung 1 - 2/3 + 5/8 - 13/21 (Deret Fibonacci)")
    print("0. Keluar")

    pilihan = input("Masukkan pilihan: ")

    if pilihan == "1":
        try:
            a = int(input("Masukkan suatu bilangan bulat: "))
            b = int(input("Masukkan pangkat yang diinginkan: "))

            for i in range(1, b + 1):
                hasil = a ** i
                print(f"{a} pangkat {i} = {hasil}")
        except ValueError:
            print("Error: Mohon masukkan angka bulat saja.")

    elif pilihan == "2":
        try:
            n = int(input("Masukkan jumlah suku (N): "))
            hasil = 0

            fibo = [1, 1]
            for i in range(2, n * 2):
                fibo.append(fibo[-1] + fibo[-2])

            for i in range(n):
                pembilang = fibo[i * 2]
                penyebut = fibo[i * 2 + 1]

                suku = pembilang / penyebut

                if i % 2 == 0:
                    hasil += suku
                    simbol = "+" if i > 0 else ""
                else:
                    hasil -= suku
                    simbol = "-"

                print(f"{simbol} ({pembilang}/{penyebut})", end=" ")

            print(f"\nHasil akhir: {hasil}")
        except ValueError:
            print("Error: Mohon masukkan angka bulat saja.")

    elif pilihan == "0":
        print("Program selesai. Sampai jumpa!")
        break

    else:
        print("Pilihan tidak valid, silakan coba lagi.")
```
